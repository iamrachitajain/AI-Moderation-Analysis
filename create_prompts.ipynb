```python
from environment import MiniWoBEnvironment
import numpy as np
import yaml
from collections import defaultdict
import dill
import pandas as pd
```

This notebook extracts prompts from miniwob environments. After some analysis and testing, only click-checkboxes-soft was deemed relevant for use in this project. 500 miniwob utterances were used.

```python
envs = {
 'click-checkboxes-soft':10,
 'click-tab-2-hard':20,
 'use-autocomplete':10,
 }
list(enumerate(envs.keys()))
```

```python
all_prompts = {}
for env_type in [x for x in envs.keys() if x not in all_prompts]:
    print(env_type)
    prompts = []
    while len(prompts)<500:
        env = MiniWoBEnvironment(env_type)
        observation, info = env.reset()
        instruction = observation['utterance']
        if instruction not in prompts:
            prompts.append(instruction)
            #print(instruction)
        if len(prompts)%10 == 0:
            print(len(prompts))
        env.close()
    all_prompts[env_type] = prompts
    dill.dump(all_prompts,open('miniwob_prompts.pkd','wb'))
all_prompts
```
